```python
assert len(files) == 9
```

```python
import os
for name, url in files:
    os.system(f"wget {url}")
    os.system(f"gunzip *.faa.gz")
    os.system(f"mv *.faa {name}.faa.bak")
```

```python
for name, _ in files:
    os.system(f"mv {name}.faa.bak {name}.faa")
```

```python
for name, _ in files:    
    print(f"makeblastdb -in {name}.faa -dbtype 'prot' -out db_{name}")
    os.system(f"makeblastdb -in {name}.faa -dbtype 'prot' -out db_{name}")
```

```python
from tqdm.auto import tqdm
import os

for i in tqdm(range(len(files)), desc="first file"):
    for j in tqdm(range(len(files)), desc="second file", leave=False):
        if i == j:
            continue
        name1, name2 = files[i][0], files[j][0]
        os.system(f"blastp -query {name1}.faa -db db_{name2} -num_threads 8 -outfmt 6 -out blast_results/{name1}_{name2}.txt")
```
